## Linear Regression

**Decide what columns to keep for X and which ones for y**

```python
df.columns
```

```python
x = df[['Avg. Session Length', 'Time on App',
       'Time on Website', 'Length of Membership']]
y=df['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split
```

**Split the data using train_test_split**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3,random_state= 101)
```

```python
len(x_train)
```

```python
len(x_test)
```

**Import the LinearRegression class and create an object of it**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

**Fit the training data into the algorithm**

```python
lr.fit(x_train, y_train)
```

**Check the co-efficients**

```python
lr.coef_
```

**Make predictions and print them**

```python
predict = lr.predict(x_test)
```

```python
predict
```
